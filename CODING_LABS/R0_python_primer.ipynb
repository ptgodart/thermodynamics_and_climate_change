### Functions

```python
# Example 1
def add_function(i, j): # Create functions. Indentation matters!
  ans = i + j
  return ans

print(add_function(1, 2))

# Example 2
add_function_inline = lambda i, j: i + j # Simple functions can be specified in 
                                         # a single line using the lambda syntax

print(add_function_inline(1, 2))

# Example 3
def complicated_function(a, b, c, d): # You can return multiple values as tuple
  x = a + b
  y = (c - d)**1.5
  z = a * b * c
  ans = 2*x + y + z**1.5
  return ans, x, y, z

print(complicated_function(1, 5, 7, 6))
print(complicated_function(1, 5, 7, 6)[0])
```

### Iterating and Loops

```python
# Logic, flow control, and loops
T_array_K = [270, 275, 270, 480, 360, 300] # K (ALWAYS PUT UNITS HERE)

# Convert T_array_K to degC
T_array_C = [] # degC
T_array_C.append(T_array_K[0] - 273.15)
T_array_C.append(T_array_K[1] - 273.15)
T_array_C.append(T_array_K[2] - 273.15)
T_array_C.append(T_array_K[3] - 273.15)
T_array_C.append(T_array_K[4] - 273.15)
T_array_C.append(T_array_K[5] - 273.15)
print(T_array_C)

# for loop:
T_array_C = [] # degC
for T in T_array_K:
  T_array_C.append(T - 273.15)
print(T_array_C)

# list comprehension:
T_array_C = [T - 273.15 for T in T_array_K] # degC
print(T_array_C)
```

## Third-party libraries

```python
import numpy as np # This shorthand allows you to write 'np' instead of 'numpy'

normal_list = [1, 2, 4, 5]
numpy_array = np.array(normal_list)
print(numpy_array)

print([i*5 for i in normal_list])
print(numpy_array * 5)

list_of_integers = range(4, 10)
print([i for i in list_of_integers])

linear_array = np.linspace(200, 400, 10)
print(linear_array)

log_array = np.logspace(0, 2, 10)
print(log_array)
```

```python
from matplotlib import pyplot as plt
import math

x = np.linspace(0, 1000, 20)
y = np.array([l**2 for l in x])

plt.plot(x, y, 'o', label='Quadratic function')
plt.xlabel('$x$')
plt.ylabel('$y = x^2$')
plt.grid()
plt.title('Linear Scaling')
plt.legend()
plt.show()

x = np.logspace(math.log10(1e-20), math.log10(1000), 20)
y = np.array([l**2 for l in x])
plt.loglog(x, y, 'o', label='Quadratic function')
plt.xlabel('$x$')
plt.ylabel('$y = x^2$')
plt.grid()
plt.title('Logarithmic Scaling')
plt.legend()
plt.show()
```

```python
from scipy.integrate import quad

def quadratic(x):
  return(x ** 2)

integral = quad(quadratic, 0, 100)
print(f"The integral of x^2 from 0 to 100 is {integral[0]:0.3g}")

x = np.linspace(-10, 10, 100)
y = np.array([l ** 2 for l in x])

plt.plot(x, y, label='y = x ** 2')
plt.plot(x, [(l ** 3) / 3 for l in x], label='y = x ** 3 / 3')
plt.plot(x, [quad(quadratic, -10, l)[0] + ((-10) ** 3) / 3 for l in x], label='quad integral')
plt.grid()
plt.legend()
plt.show()
```

```python
import CoolProp as cp
from CoolProp.CoolProp import AbstractState

Water = AbstractState('HEOS', 'Water')  # The first argument selects the type of 
                                        # model (e.g. 'HEOS' = Helmholtz Equation 
                                        # of State) and the second argument 
                                        # selects the species (e.g. 'Water')

P_0 = 101325 # Pa
T_0 = 300 # K
Water.update(cp.PT_INPUTS, P_0, T_0)
print(f'The entropy of Water at {P_0:.3g} Pa and {T_0:.3g} K is {Water.smass():.3g} J/kg-K or {Water.smolar():.3g} J/mol-K')
```

```python
CO2 = AbstractState('HEOS', 'CarbonDioxide')
T_vec = np.linspace(300, 1000, 250) # K
P_0 = 101325 # Pa

u_vec = [] # J/kg
for T in T_vec:
  CO2.update(cp.PT_INPUTS, P_0, T)
  u_vec.append(CO2.umass())

plt.plot(T_vec, u_vec)
plt.xlabel('$T$ [K]')
plt.ylabel('$u$ [J/kg]')
plt.show()
```

## Entropy Example

Let's implement the following equation:

$$\Delta s = c_v\ln\left(\frac{T_2}{T_1}\right) + R\ln\left(\frac{v_2}{v_1}\right)$$

*Note: you can input $\LaTeX$-formatted equations directly into these text cells.*

```python
c_v_CO2 = 29 # J/mol-K
CO2.update(cp.PT_INPUTS, 101325, 300.15)
print(CO2.cvmolar())
R = 8.314 # J/mol-K

T_1 = 300 # K
T_2 = 600 # K
v_1 = 3 # m^3/mol
v_2 = 2 # m^3/mol
delta_s = c_v_CO2*math.log(T_2/T_1) + R*math.log(v_2/v_1)
print(f'delta_s = {delta_s:.3g} J/mol-K')

delta_s_func = lambda T: c_v_CO2*math.log(T/T_1) + R*math.log(v_2/v_1)
T_2_vec = np.linspace(300, 1000, 250)
delta_s_vec = [delta_s_func(T) for T in T_2_vec]
plt.plot(T_2_vec, delta_s_vec)
plt.xlabel('$T_2$ [K]')
plt.ylabel('$\Delta s$ [J/mol-K]')
plt.show()
```
